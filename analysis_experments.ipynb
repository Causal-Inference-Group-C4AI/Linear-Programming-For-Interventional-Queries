```python
import pandas as pd
import numpy as np
experiments_result_path = "new.csv" 
df = pd.read_csv(experiments_result_path)
df['GC_LOWER_DIFF_BOUND'] = df['GC_LOWER_BOUND'] - df['LP_LOWER_BOUND']
df['GC_UPPER_DIFF_BOUND'] = df['GC_UPPER_BOUND'] - df['LP_UPPER_BOUND']
df['GC_BOUND_SIZE'] = df['GC_UPPER_BOUND'] - df['GC_LOWER_BOUND']
df['LP_BOUND_SIZE'] = df['LP_UPPER_BOUND'] - df['LP_LOWER_BOUND']
df['LOWER_DIFF_BOUND'] = df['GC_LOWER_BOUND'] - df['LP_LOWER_BOUND']
df['UPPER_DIFF_BOUND'] = df['GC_UPPER_BOUND'] - df['LP_UPPER_BOUND']
df['BOUND_CHECK'] = ((df['TRUE_VALUE'] >= df['GC_LOWER_BOUND']) & (df['TRUE_VALUE'] <= df['GC_UPPER_BOUND']) & (df['TRUE_VALUE'] >= df['LP_LOWER_BOUND']) & (df['TRUE_VALUE'] <= df['LP_UPPER_BOUND']))
df['BOUND_CHECK'] = np.where(
    (df['GC_LOWER_BOUND'].isna() | df['GC_UPPER_BOUND'].isna()),
    True,
    ((df['TRUE_VALUE'] >= df['GC_LOWER_BOUND']) & (df['TRUE_VALUE'] <= df['GC_UPPER_BOUND']) &
    (df['TRUE_VALUE'] >= df['LP_LOWER_BOUND']) & (df['TRUE_VALUE'] <= df['LP_UPPER_BOUND']))
)


```

```python
data_df = pd.DataFrame(columns=['N', 'M', 'GC_LOWER_ITERATIONS_MEAN', 'GC_LOWER_ITERATIONS_SD', 'GC_UPPER_ITERATIONS_MEAN', 'GC_UPPER_ITERATIONS_SD', 'GC_BOUND_SIZE_MEAN', 'GC_BOUND_SIZE_SD', 'GC_TIME_MEAN','GC_TIME_SD', 'LP_BOUND_SIZE_MEAN', 'LP_BOUND_SIZE_SD', 'LP_TIME_MEAN', 'LP_TIME_SD', 'LOWER_DIFF_MEAN', 'LOWER_DIFF_SD', 'UPPER_DIFF_MEAN', 'UPPER_DIFF_SD'])


colunas = [
    ('GC_LOWER_BOUND_REQUIRED_ITERATIONS', 'GC_LOWER_ITERATIONS_'),
    ('GC_UPPER_BOUND_REQUIRED_ITERATIONS', 'GC_UPPER_ITERATIONS_'),
    ('GC_BOUND_SIZE','GC_BOUND_SIZE_'),
    ('LP_BOUND_SIZE','LP_BOUND_SIZE_'),
    ('GC_SECONDS_TAKEN','GC_TIME_'),
    ('LP_SECONDS_TAKEN','LP_TIME_'),
    ('LOWER_DIFF_BOUND', 'LOWER_DIFF_'),
    ('UPPER_DIFF_BOUND', 'UPPER_DIFF_'),
]

N_M = [
        (1,1),
        (2,1),
        (1,2),
        (3,1),
        (4,1),
        (2,2),
        #(1,3),
        #(2,3), # Not LP
]
for n_m in N_M:
    N, M = n_m
    new_row = {'N': N, 'M': M, 'GC_LOWER_ITERATIONS_MEAN': None, 'GC_LOWER_ITERATIONS_SD': None, 'GC_UPPER_ITERATIONS_MEAN': None, 'GC_UPPER_ITERATIONS_SD': None, 'GC_BOUND_SIZE_MEAN': None, 'GC_BOUND_SIZE_SD': None, 'GC_TIME_MEAN': None,'GC_TIME_SD': None, 'LP_BOUND_SIZE_MEAN': None, 'LP_BOUND_SIZE_SD': None, 'LP_TIME_MEAN': None, 'LP_TIME_SD': None, 'LOWER_DIFF_MEAN': None, 'LOWER_DIFF_SD': None, 'UPPER_DIFF_MEAN': None, 'UPPER_DIFF_SD': None}
    new_row_df = pd.DataFrame([new_row])
    for coluna in colunas:
        c1, c2 = coluna
        filtered_data = df[(df['N'].eq(N)) & (df['M'].eq(M)) & (df[c1].notna())]

        mean_val = filtered_data[c1].mean() if not filtered_data.empty else np.nan
        std_val = filtered_data[c1].std(ddof=0) if len(filtered_data) > 1 else 0

        new_row_df[f'{c2}MEAN'] = mean_val
        new_row_df[f'{c2}SD'] = std_val

    filtered_data = df[(df['N'].eq(N)) & (df['M'].eq(M)) & (df['BOUND_CHECK'].notna())]
    all_true = filtered_data['BOUND_CHECK'].all()
    new_row_df['ALL_CHECKED'] = all_true
    data_df = pd.concat([data_df, new_row_df], ignore_index=True)

```
