```python
import pandas as pd

df_algorithm = pd.read_csv("algorithm_results.csv")
df_column = pd.read_csv("results.csv")
df_true = pd.read_csv("true_results.csv")
df_new_server_results = pd.read_csv("new_results_server.csv")

df_merged = pd.merge(df_algorithm, df_column, on=['N', 'M'], how='outer')
df_merged = pd.merge(df_merged, df_true, on=['N', 'M'], how='outer')
# print(df_merged.head())
df_merged['difference'] = df_merged['BOUNDS_SIZE'] - df_merged['ALGO_BOUNDS_SIZE']
df_merged['difference']
df_merged.to_csv("merged_bounds.csv", index=False)

df = pd.DataFrame()
# df['COLUMN_GEN'] = True if isinstance(df_merged['LOWER_BOUND'], float) else False
# df['ALGORITHM'] = True if isinstance(df_merged['LOWER_BOUND'], float) else False
df['N'] = df_merged['N']
df['M'] = df_merged['M']
df['COLUMN_GEN'] = df_merged['LOWER_BOUND'].notna()
df['ALGORITHM'] = df_merged['ALGO_LOWER_BOUND'].notna()
df['CG_LOWER'] = df_merged['LOWER_BOUND']
df['CG_UPPER'] = df_merged['UPPER_BOUND']
df['DIFF_LOWER'] = df_merged['LOWER_BOUND']-df_merged['ALGO_LOWER_BOUND']
df['DIFF_UPPER'] = df_merged['UPPER_BOUND']-df_merged['ALGO_UPPER_BOUND']
df['TRUE_VALUE'] = df_merged['TRUE_VALUE']
df['CG_TIME'] = df_merged['LOWER_BOUND_SECONDS_TAKEN']+df_merged['UPPER_BOUND_SECONDS_TAKEN']
df['ALGORITHM_TIME'] = df_merged['ALGO_BOUNDS_SECONDS_TAKEN']
df['CG_LOWER_Ite'] = df_merged['LOWER_BOUND_REQUIRED_ITERATIONS']
df['CG_UPPER_Ite'] = df_merged['UPPER_BOUND_REQUIRED_ITERATIONS'].astype(int)
df['CG_BOUNDS_SIZE'] = df_merged['BOUNDS_SIZE']
```
