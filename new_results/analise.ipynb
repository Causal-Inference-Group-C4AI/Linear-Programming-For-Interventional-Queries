```python
import pandas as pd

df_algorithm = pd.read_csv("algo.csv")
df_true = pd.read_csv("true_results.csv")
df_gc = pd.read_csv("gc.csv")

df_merged = pd.merge(df_gc, df_algorithm, on=['N', 'M'], how='outer')
df_merged = pd.merge(df_merged, df_true, on=['N', 'M'], how='inner')


df = pd.DataFrame()
df['N'] = df_merged['N']
df['M'] = df_merged['M']
df['COLUMN_GEN'] = df_merged['LOWER_BOUND'].notna()
df['ALGORITHM'] = df_merged['ALGO_LOWER_BOUND'].notna()
df['CG_LOWER'] = df_merged['LOWER_BOUND']
df['CG_UPPER'] = df_merged['UPPER_BOUND']
df['DIFF_LOWER'] = df_merged['LOWER_BOUND']-df_merged['ALGO_LOWER_BOUND']
df['DIFF_UPPER'] = df_merged['UPPER_BOUND']-df_merged['ALGO_UPPER_BOUND']
df['TRUE_VALUE'] = df_merged['TRUE_VALUE']
df['CG_TIME'] = df_merged['TOTAL_SECONDS_TAKEN']
df['ALGORITHM_TIME'] = df_merged['ALGO_TOTAL_SECONDS_TAKEN']
df['CG_LOWER_Ite'] = df_merged['LOWER_BOUND_REQUIRED_ITERATIONS'].astype('int64')
df['CG_UPPER_Ite'] = df_merged['UPPER_BOUND_REQUIRED_ITERATIONS'].astype('int64')
df.to_csv("latest_results.csv", index=False)
```
