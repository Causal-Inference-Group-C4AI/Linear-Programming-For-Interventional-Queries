```python
import pandas as pd

df_algorithm = pd.read_csv("intermediario_algorithm_results.csv")
df_column = pd.read_csv("results.csv")
df_true = pd.read_csv("true_results.csv")

df_merged = pd.merge(df_algorithm, df_column, on=['N', 'M'], how='inner')
df_merged = pd.merge(df_merged, df_true, on=['N', 'M'], how='inner')
df_merged.to_csv("merged_bounds.csv", index=False)
# print(df_merged.head())
df_merged['difference'] = df_merged['ALGO_BOUNDS_SIZE'] - df_merged['BOUNDS_SIZE']
df_merged['difference']

df = pd.DataFrame()
df['UPPER'] = df_merged['UPPER_BOUND']
df['LOWER'] = df_merged['LOWER_BOUND']
df['TRUE'] = df_merged['TRUE_VALUE']
df['N'] = df_merged['N']
df['M'] = df_merged['M']
```
